```python
import pandas as pd
import matplotlib.pyplot as plt

# Compute correlation with 'charges'
corr_with_charges = df.corr(numeric_only=True)['charges'].sort_values()

# Plot
plt.figure(figsize=(12, 8))  # adjust size for spacing since there are ~22 features
corr_with_charges.plot(kind='barh', color='skyblue', edgecolor='black')

plt.title('Correlation of Features with Charges', fontsize=14)
plt.xlabel('Correlation Coefficient', fontsize=12)
plt.ylabel('Features', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
df.drop('charges',axis=1)
```

```python
x = df.drop('charges',axis=1)
y = df['charges']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
prediction = lr.predict(x_test)
```

```python
prediction
```

```python
import matplotlib.pyplot as plt

# Predict the values
prediction = lr.predict(x_test)

# Create an index for x-axis
x_axis = range(len(y_test))

# Plot
plt.figure(figsize=(12, 6))
plt.scatter(x_axis, y_test, color='blue', label='Actual', alpha=0.6)
plt.scatter(x_axis, prediction, color='red', label='Predicted', alpha=0.6)

# Labels and title
plt.title('Actual vs Predicted Charges', fontsize=14)
plt.xlabel('Sample Index', fontsize=12)
plt.ylabel('Charges', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_percentage_error,root_mean_squared_error,r2_score





```

```python
mean_absolute_percentage_error(y_test, prediction) * 100
```

```python
root_mean_squared_error(y_test,prediction)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Initialize the Decision Tree Regressor
dt_model = DecisionTreeRegressor(random_state=42)

# Train the model
dt_model.fit(x_train, y_train)
```

```python
# Make predictions on the test set
dt_prediction = dt_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error, root_mean_squared_error, r2_score

# Evaluate the model
dt_mape = mean_absolute_percentage_error(y_test, dt_prediction) * 100
dt_rmse = root_mean_squared_error(y_test, dt_prediction)
dt_r2 = r2_score(y_test, dt_prediction)

print(f"Decision Tree Regressor Performance:")
print(f"Mean Absolute Percentage Error (MAPE): {dt_mape:.2f}")
print(f"Root Mean Squared Error (RMSE): {dt_rmse:.2f}")
print(f"R-squared (R2): {dt_r2:.2f}")
```

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize the Random Forest Regressor
rf_model = RandomForestRegressor(random_state=42)

# Train the model
rf_model.fit(x_train, y_train)
```

```python
# Make predictions on the test set
rf_prediction = rf_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error, root_mean_squared_error, r2_score

# Evaluate the model
rf_mape = mean_absolute_percentage_error(y_test, rf_prediction) * 100
rf_rmse = root_mean_squared_error(y_test, rf_prediction)
rf_r2 = r2_score(y_test, rf_prediction)

print(f"Random Forest Regressor Performance:")
print(f"Mean Absolute Percentage Error (MAPE): {rf_mape:.2f}")
print(f"Root Mean Squared Error (RMSE): {rf_rmse:.2f}")
print(f"R-squared (R2): {rf_r2:.2f}")
```
